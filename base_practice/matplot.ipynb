```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

def f(x):
    return (x - 2) * x * (x + 2)
```

```python
# ndarrayに対してfを呼び出すと、それぞれの要素ごとに関数が適用される
print(f(np.array([1, 2, 3])))
```

```python
x = np.arange(-3, 3.5, 0.5)
print(x)
```

```python
# グラフ描画の際には、linspaceが便利
x = np.linspace(-3, 3, 10)
print(np.round(x, 2)) # roundで小数点以下第2位に四捨五入
```

```python
plt.plot(x, f(x))
plt.show()
```

```python
# グラフをもうちょっと見栄えよくしてみる
def f2(x, w):
     return (x - w) * x * (x + w)

x = np.linspace(-3, 3, 100)

plt.plot(x, f2(x, 1), color='black', label='$w=1$') # $$で囲むとtex形式の数式として指定できる
plt.plot(x, f2(x, 2), color='cornflowerblue', label='$w=2$')
plt.legend(loc='best') # 数字や"upper left"、"best"などで指定できる
plt.ylim(-15, 15)
plt.title('$f_2(x)$')
plt.xlabel('$x$')
plt.ylabel('$y$')
plt.grid(True)
plt.show()
```

```python
# グラフの色の種類
import matplotlib
matplotlib.colors.cnames
```

```python
# グラフを複数並べる
plt.figure(figsize=(10, 3)) # 描画領域を指定
plt.subplots_adjust(wspace=0.5, hspace=0.5) # グラフの間隔を指定
for i in range(6):
    plt.subplot(2, 3, i + 1) # グラフ描画の位置を指定
    plt.title(i + 1)
    plt.plot(x, f2(x, i), 'k')
    plt.ylim(-20, 20)
    plt.grid(True)
plt.show()
```

```python
# 他にもいろいろな書き方がある
fig, axes = plt.subplots(2, 3, figsize=(10, 3))
ax = axes.ravel()
for i in range(6):
    ax[i].plot(x, f2(x, i)) 
    ax[i].set_title(i + 1)
    ax[i].set_ylim(-20, 20)
    ax[i].grid(True)
fig.tight_layout()
plt.show()
```

```python
# 3次元のグラフを描いてみる

def f3(x0, x1):
    ans = (2 * x0**2 + x1**2) * np.exp(-(2 * x0**2 + x1**2))
    return ans

xn = 9
x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)
y = np.zeros((len(x0), len(x1)))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])
```

```python
print(x0)
```

```python
print(np.round(y, 1))
```

```python
# 2次元行列の要素を色に置き換えて表示
plt.figure(figsize=(3.5, 3))
plt.gray() # グレースケールで表す
plt.pcolor(y) # ←ここをいろいろな関数にすると表示を変えられる
plt.colorbar()
plt.show()
```

```python
plt.figure(figsize=(3.5, 3))
plt.gray() # グレースケールで表す
plt.contour(y)
plt.colorbar()
plt.show()
```

```python
# 面で関数を表す
from mpl_toolkits.mplot3d import Axes3D

xx0, xx1 = np.meshgrid(x0, x1) # 座標点を作成

plt.figure(figsize=(5, 3.5))
ax = plt.subplot(1, 1, 1, projection='3d') # projectionで3次元表示に (row, col, n)で図をrow行col列に分割して、左上からn番目のところに描画する
ax.plot_surface(xx0, xx1, y, rstride=1, cstride=1, alpha=.3, color='blue', edgecolor='black') 
# rstride, cstrice 縦と横でどれぐらいの要素ずつ線を引くかを指定できる。小さいほど間隔が密になる
ax.set_zticks((0, 0.2)) # z軸のメモリを0と0.2だけにする
ax.view_init(75, -95) # 0→真横、90→真上
plt.show()
# 3次元表示はpractice-2.ipynbのサポートベクタマシンのところで出てくる
```

```python
# 等高線で表示
xn = 50
x0 = np.linspace(-2, 2, xn)
x1 = np.linspace(-2, 2, xn)

y = np.zeros((xn, xn))
for i0 in range(xn):
    for i1 in range(xn):
        y[i1, i0] = f3(x0[i0], x1[i1])

xx0, xx1 = np.meshgrid(x0, x1)

plt.figure(1, figsize=(4, 4))
cont = plt.contour(xx0, xx1, y, 5, colors='black') # 5=表示する高さのレベルを5段階に設定
cont.clabel(fmt='%.2f', fontsize=8)
plt.xlabel('$x_0$', fontsize=14)
plt.ylabel('$x_1$', fontsize=14)
plt.show()
```
